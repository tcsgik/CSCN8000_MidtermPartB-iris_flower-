## 2.	Use the iris flower dataset from the sklearn library and cross_val_score against the logistic regression model to measure the performance of each. 

### Step 1: load dataset and define features and target variables

```python
# Import libraries
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
import numpy as np

# Load iris flower dataset
iris = load_iris()

# Define features and target variables
X = iris.data
y = iris.target
```

### Step 3: logistic regression model and cross-validation

```python
# Create a logistic regression model
logreg = LogisticRegression(max_iter=200)

# Perform cross-validation
scores = cross_val_score(logreg, X, y, cv=5) # 5-fold cross validation

# Display the cross-validation scores and their average
print("Cross-validation scores:", scores)
print("Average cross-validation score: {:.2f}".format(scores.mean()))
```

```python
# Import library
import matplotlib.pyplot as plt

# Visualize the cross-validation scores
plt.boxplot(scores, vert=False, patch_artist=True)
plt.title('Cross-Validation Scores (Logistic Regression on Iris)')
plt.xlabel('Accuracy')
plt.yticks([1], ['Logistic Regression'])
plt.grid(axis='x')
plt.show()
```

### Observations and reflections:
- Most cross-validation scores are above 0.90, with an average of about 0.96, indicating that the model performs very well in the classification of the iris dataset.
- The scores of each fold fluctuate slightly because the data distribution is slightly different each time the training/test set is split, but the overall stability is stable.
- Cross-validation can more comprehensively evaluate the generalization ability of the model and avoid the contingency caused by relying only on a single training/test split.

### Conclusion:
- Using cross_val_score with Logistic Regression can quickly evaluate the overall performance of the model in the iris dataset, with an average accuracy of up to 96%.
